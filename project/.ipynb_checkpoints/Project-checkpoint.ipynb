```python
from netCDF4 import Dataset
import numpy as np
from metpy.plots import wx_code_to_numeric as wx2num
import pandas as pd
from metpy.units import units
import metpy.calc as mpcalc
```

```python
data = Dataset("./20130502_1800")
```

```python
station_name = data.variables["stationName"][:]
#print(station_name)

#for var in data.variables:
#    print(var)

#Read in variables as needed

latitude = data.variables["latitude"][:]
longitude = data.variables["latitude"][:]

#current weather
curweather = data.variables["presWeather"][:]
#temperature
t = data.variables["temperature"][:]
#dewpoint
dwp = data.variables["dewpoint"][:]
#winds (windSpeed)
windspd = data.variables["windSpeed"][:]
#wind direction(windDir)
winddir = data.variables["windDir"][:]
#gusts (windGust)
gusts = data.variables["windGust"][:]
#station name (station_name)
#pressure(seaLevelPres)
pres = data.variables["seaLevelPress"][:]
#visibility(visitbility)
vis = data.variables["visibility"][:]
#skycover
skycover = data.variables["skyCover"][:]
```

```python
print(curweather)
```

```python
stn = ['']*len(station_name)
#stn0 = [station_name[0].tobytes().decode().rstrip('\x00'),station_name[1].tobytes().decode()]
#.tobytes() changes bytes masked to byte array
#.decode() changes byte array to string
#.rstrip() removes whatever we put into the function
for index in range(len(station_name)):
    stn[index] = station_name[index].tobytes().decode().rstrip('\x00')

print(stn)
```

```python
print(skycover)
```

```python
sky_cover_num = np.empty(len(skycover))
for i in range(len(skycover)):
    update_sky = skycover[i].tobytes().decode().rstrip('\x00')
    #if skc = 0/8
    if update_sky[-3:]== 'SKC' or update_sky[-3:] == 'CLR':
        sky_cover_num[i] = 0
    #if skc = 2/8
    if update_sky[-3:]== 'FEW':
        sky_cover_num[i] = 2
    #if skc = 4/8
    if update_sky[-3:]== 'SCT':
        sky_cover_num[i] = 4
    #if skc = 6/8
    if update_sky[-3:]== 'BNK':
        sky_cover_num[i] = 6    
    #if skc = 8/8
    if update_sky[-3:]== 'OVC':
        sky_cover_num[i] = 8
    else:
        sky_cover_num[i] = 10

print(update_sky)
```
